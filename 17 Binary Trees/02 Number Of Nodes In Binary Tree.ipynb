```python
class BinaryTree:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None
```

```python
def BinaryTreeInput():
  rootData = int(input())
  if(rootData == -1):
    return None
  root = BinaryTree(rootData)
  root.left = BinaryTreeInput()
  root.right = BinaryTreeInput()
  return root
```

```python
def PrintBinaryTreeDetailed(root):
  if(root == None):
    return
  print(root.data, end = " : ")
  if(root.left != None):
    print("L = ", root.left.data, end = ", ")
  if(root.right != None):
    print("R = ", root.right.data, end = "")
  print()
  PrintBinaryTreeDetailed(root.left)
  PrintBinaryTreeDetailed(root.right)
```

```python
def numberOfNodes(root):
  if(root == None):
    return 0
  return 1 + numberOfNodes(root.left) + numberOfNodes(root.right)
```

```python
root = BinaryTreeInput()
PrintBinaryTreeDetailed(root)
print(numberOfNodes(root))
```
