```python
class MapNode:
  def __init__(self, key, value):
    self.key = key
    self.value = value
    self.next = None

class Map:
  def __init__(self):
    self.bucketSize = 5
    self.bucket = [None] * self.bucketSize
    self.count = 0

  def size(self):
    return self.count

  def compressionFunction(self, hashCode):
    return abs(hashCode) % (self.bucketSize)

  def rehash(self):
    temp = self.bucket
    self.bucket = [None] * (2 * self.bucketSize)
    self.bucketSize = 2 * self.bucketSize
    self.count = 0
    for head in temp:
      while(head != None):
        self.insert(head.key, head.value)
        head = head.next

  def loadFactorPrinter(self):
    return self.count / self.bucketSize

  def insert(self, key, value):
    hashCode = hash(key)
    index = self.compressionFunction(hashCode)
    head = self.bucket[index]
    while head != None:
      if(head.key == key):
        head.value = value
        return
      head = head.next
    head = self.bucket[index]
    newNode = MapNode(key, value)
    newNode.next = head
    self.count += 1
    self.bucket[index] = newNode
    loadFactor = self.count / self.bucketSize
    if(loadFactor >= 0.7):
      self.rehash()

  def search(self, key):
    hashCode = hash(key)
    index = self.compressionFunction(hashCode)
    head = self.bucket[index]
    while(head != None):
      if(head.key == key):
        return head.value
      head = head.next
    return False

  def delete(self, key):
    hashCode = hash(key)
    index = self.compressionFunction(hashCode)
    head = self.bucket[index]
    prev = None
    while(head != None):
      if(head.key == key):
        if(prev == None):
          self.bucket[index] = None
        else:
          prev.next = head.next
        self.count -= 1
        return head.value
      prev = head
      head = head.next
    return "The Key Is Not Present"

m = Map()
for i in range(10):
  m.insert("abc" + str(i), i + 1)
  print(m.loadFactorPrinter())
print()
for i in range(10):
  print("abc" + str(i) + " : ", (m.search("abc" + str(i))))


# m.insert("Harsh", "Prasad")
# print(m.size())
# m.insert("Ayush", 12)
# print(m.size())
# m.insert("Harsh", 25)
# print(m.size())
# print(m.search("harsh"))
# print(m.search("Ayush"))
# print(m.search("Harsh"))
# print()
# print(m.delete("harsh"))
# print(m.delete("Ayush"))
# print(m.delete("Ayush"))
# print(m.delete("Harsh"))
# print(m.delete("Harsh"))
```
