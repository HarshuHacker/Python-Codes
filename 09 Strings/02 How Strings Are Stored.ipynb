As We Have Seen In Integers, For Small Integers (-5 to 256) The Address For These Numbers Were Shared.
Similarly In Strings, The Address Of Small Strings Are Shared.

```python
a = "Harsh"
print(id(a))
b = "Harsh"
print(id(b))
c = "Hardh"
print(id(c))
d = ""
for i in c:
  if(i == "d"):
    d += "s"
    continue
  d += i
print(id(d))
print(a)
print(b)
print(c)
print(d)
```
