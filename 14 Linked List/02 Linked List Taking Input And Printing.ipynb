```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None

def takeInput():
  inputList = [int(x) for x in input().split()]
  head = None
  for currData in inputList:
    if(currData == -1):
      break
    newNode = Node(currData)
    if(head == None):
      head = newNode
      tail = newNode
    else:
      tail.next = newNode
      tail = newNode
  return head

head = takeInput()
while(head != None):
  print(head.data, "->", end = " ")
  head = head.next
print("None")
```
